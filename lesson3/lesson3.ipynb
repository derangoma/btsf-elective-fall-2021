```python
# recap of the anatomy of a function

# this function returns -1 if the argument `x` is negative, and 2x otherwise
def f(x):
    if x < 0:
        return -1
    else:
        return 2 * x
```

```python
# examples on various arguments
print(f(-100))
print(f(-10))
print(f(0))
print(f(5))
print(f(100))
```
